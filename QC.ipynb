# Remove features with low variance

```python
%%R
# identify variance for each column
varCol=apply(num.mat, 2, var, na.rm = T)
```

```python
%%R
varCol[1:2]
```

```python
%%R
# identify columns with zero vairance i.e (constant among all samples)
constCol <- (varCol == 0 | is.na(varCol))
constCol[1:10]
```

```python
%%R
# 31 columns/ features with zero variance
sum(constCol)
```

```python
%%R
# remove these columns
num.mat <- num.mat[, !constCol]
#head(num.mat)
```

# compute missingness rate and impute missing values


Missing value estimation methods for DNA microarrays
https://academic.oup.com/bioinformatics/article/17/6/520/272365?login=true

https://scholar.google.com/citations?user=PjCDBjIAAAAJ&hl=en

```python
%%R
# for each feature/column how many missing values divided by number of samples
#apply(is.na(num.mat), 2, sum)
#nrow(num.mat)


missingness_rate=apply(is.na(num.mat), 2, sum)/nrow(num.mat) *100
missingness_rate[1:5]
```

```python
%%R
min(missingness_rate)
```

```python
%%R
max(missingness_rate)
```

```python
%%R
h=hist(missingness_rate,breaks=10,
main="",
xlab="percentage of missingness")
text(h$mids,h$counts,labels=h$counts, adj=c(0.5, -0.5))
```

```python
%%R
good.inx=missingness_rate < 50
good.inx[1:10]
```

```python
%%R
num.mat=num.mat[,good.inx]
dim(num.mat)
#head(num.mat)
```
